📝 Redacción del problema
Este programa interactivo permite calcular el área y el perímetro de un rectángulo. El usuario selecciona una opción del menú, ingresa la base y la altura cuando se le solicita, y el programa muestra el resultado correspondiente. El menú se repite hasta que el usuario decida salir.


* 📐 Fórmulas a utilizar y su explicación
Concepto
Fórmula
Explicación sencilla
* Área
A = b × h
Multiplicamos la base (b) por la altura (h). Nos dice cuántas unidades cuadradas "caben" dentro del rectángulo.
* Perímetro
P = (b+h+b+h)
Sumamos base + altura y multiplicamos por 2. Representa la distancia total del borde que rodea al rectángulo.

![texto del vinculo](https://raw.githubusercontent.com/alanromero128/Mi-imagenes-de-EPPL-Unidad-2/refs/heads/main/rectangulo%20area%20y%20perimetro.jpg)

El programa está organizado en tres funciones que colaboran entre sí. area_rectangulo() y perimetro_rectangulo() son las encargadas de pedir la base y la altura al usuario, aplicar sus fórmulas matemáticas y mostrar el resultado. La función menu() actúa como controlador principal: usa un bucle while True para mantener el menú visible de forma continua. Dentro de ese ciclo, las estructuras if, elif y else leen la opción que escribe el usuario y dirigen el flujo hacia la función correspondiente. Si elige calcular, el programa solicita los datos, imprime la respuesta y vuelve automáticamente al menú; si elige salir, la palabra clave break rompe el bucle y el programa finaliza. Cualquier texto que no sea 1, 2 o 3 activa un aviso de error y reinicia el ciclo.

```python
# --- FUNCIÓN PARA CALCULAR EL ÁREA ---
def area_rectangulo():
    print("\n📏 CALCULAR ÁREA DEL RECTÁNGULO")
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))

    # Fórmula: A = b * h
    area = base * altura
    print(f"✅ El área es: {area} unidades cuadradas")

# --- FUNCIÓN PARA CALCULAR EL PERÍMETRO ---
def perimetro_rectangulo():
    print("\n📐 CALCULAR PERÍMETRO DEL RECTÁNGULO")
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))

    # Fórmula: P = 2 * (b + h)
    perimetro = 2 * (base + altura)
    print(f"✅ El perímetro es: {perimetro} unidades lineales")

# --- FUNCIÓN MENÚ (INTERACTIVA) ---
def menu():
    # while True crea un ciclo infinito que se rompe solo cuando el usuario elige salir
    while True:
        print("\n🔹 MENÚ PRINCIPAL")
        print("1. Calcular área del rectángulo")
        print("2. Calcular perímetro del rectángulo")
        print("3. Salir")

        opcion = input("Elige una opción (1, 2 o 3): ")

        # if / elif / else controlan qué función se ejecuta según la elección
        if opcion == "1":
            area_rectangulo()          # Llamada a la función de área
        elif opcion == "2":
            perimetro_rectangulo()     # Llamada a la función de perímetro
        elif opcion == "3":
            print("👋 ¡Gracias por usar el programa! Adiós.")
            break                      # Rompe el ciclo while y termina el programa
        else:
            print("⚠️ Opción no válida. Por favor, escribe 1, 2 o 3.")

# --- PUNTO DE ENTRADA DEL PROGRAMA ---
if __name__ == "__main__":
    menu()
```
